```python
# Importando as bibliotecas que vamos usar
import sqlite3
import pandas as pd

# 1. CONEXÃO: Apontamos para o nosso arquivo de banco de dados
# Como ele está na mesma pasta, só o nome já basta.
conexao = sqlite3.connect('meu_banco_de_dados.db')

# 2. A PERGUNTA (QUERY): Escrevemos nossa consulta SQL
# Esta é a mesma query com JOIN que você já dominou
query = """
SELECT
    l.timestamp,
    c.nome_produto,
    l.temperatura_secador,
    l.velocidade_rolo_m_min,
    l.umidade_papel_percent
FROM
    log_producao AS l
JOIN
    cadastro_produtos AS c ON l.codigo_produto = c.codigo_produto;
"""

# 3. A EXECUÇÃO E CARGA: A mágica do Pandas
# Usamos a função read_sql_query para enviar a query através da conexão
# e receber um DataFrame como resposta.
df_final = pd.read_sql_query(query, conexao)

# 4. BOA PRÁTICA: Fechamos a conexão com o banco de dados
conexao.close()


# --- VERIFICAÇÃO FINAL ---
print("--- DADOS CARREGADOS DO BANCO DE DADOS PARA O PANDAS ---")
print(df_final.head()) # Mostra as 5 primeiras linhas do resultado

print("\n--- INFORMAÇÕES DO DATAFRAME CRIADO ---")
print(df_final.info()) # Mostra o resumo do DataFrame
```
